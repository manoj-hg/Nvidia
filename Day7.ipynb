```python
import os
import cv2
import matplotlib.pyplot as plt

# 1. Define the absolute paths using raw strings
path1 = r"C:\Users\Student\Downloads\pexels-kerber-2708981.jpg"
path2 = r"C:\Users\Student\Downloads\pexels-suki-lee-110686949-14826118.jpg"

# 2. Pre-check: Verify if Windows can actually see the files
if not os.path.exists(path1):
    raise FileNotFoundError(f"System cannot find Image 1 at: {path1}\nCheck spelling or file extension (.jpeg?)")
if not os.path.exists(path2):
    raise FileNotFoundError(f"System cannot find Image 2 at: {path2}\nCheck spelling or file extension (.jpeg?)")

# 3. Read images in grayscale (0)
img1 = cv2.imread(path1, 0)
img2 = cv2.imread(path2, 0)

# Double check that OpenCV successfully decoded the images
if img1 is None or img2 is None:
    raise RuntimeError("OpenCV failed to decode the images. The files might be corrupted or in an unsupported format.")

# 4. Initialize ORB detector
orb = cv2.ORB_create()

# Find the keypoints and descriptors
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

# Ensure descriptors were found before matching
if des1 is None or des2 is None:
    raise ValueError("Could not find enough features in one of the images to perform matching.")

# 5. Initialize Brute-Force Matcher with Hamming distance
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors and sort them by distance (best matches first)
matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)

# 6. Draw top 20 matches (using 'thickness' instead of 'matchesThickness')
result = cv2.drawMatches(img1, kp1, img2, kp2, matches[:20], None, thickness=2)

# 7. Convert BGR to RGB for correct Matplotlib rendering and display
plt.figure(figsize=(12, 6))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Top 20 Feature Matches")
plt.axis('off')
plt.show()
```
